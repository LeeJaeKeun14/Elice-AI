# 자바 스크립트

## 자바스크립트

이미지 슬라이드 효과, 팝업 효과 등의 기능을 포함한 동적인 웹사이트 제작 시 사용되는 프로그래밍 언어

## 자바스크립트의 변수

- 변수 선언
  - 데이터를 담을 공간을 생성하는 것
- 변수 초기화
  - 생성된 변수에 데이터를 전달하는 것

### 변수 안의 데이터 확인 방법

```js
    var fruit = "apple";

    console.log(fruit);
```

- `console.log`는 변수 안에 데이터를 확인할 때 사용하는 명령어

## 데이터 타입

- String : 문자열
- Number : 숫자
- Function : 함수
- Array : 배열
- Object : 객체
- Boolean : 불린
- undefined : 정의되지 않음
- null : 널

### 함수

```js
    var area = function(width, height){
        return width * height;
    }
```
area(10, 20)


- 매개변수 : 인자로부터 전달받은 값이 들어가는 통로, 상황에 따라 생략 가능
- 인자 : 함수에게 전달하는 데이터

### 객체

```js
    var student = {
        name : "inkwon",
        age : 20,
        skills : ["자바", "HTML"],
        sum : function(num1, num2){
          return num1 + num2;
        }
    }

    console.log(student.name);
    console.log(student["name"])
```

- 프로퍼티, 메서드, 데이터로 구성
  - 변수, 함수

## 프로퍼티와 메서드

### 문자열 프로퍼티와 메서드

- .length : 문자열 길이
- .charAt(0) : 문자 H 추출
- .split(" ") : 공백 기준으로 문자를 나눈후 배열로 출력

### 배열 프로퍼티와 메서드

- .length : 대이터 개수
- .push(data) : 배열 뒤에 데이터 삽입
- .unshift(data) : 배열 앞에 데이터 삽입

- .pop() : 배열 뒤의 데이터 제거
- .shift() : 배열 앞의 데이터 제거

### math의 수학 연산 메서드

- Math.abs(-3) : 절대값
- Math.ceil(0.3) : 올림
- Math.floor(0.9) : 내림
- Math.random() : 랜덤

### 문자를 숫자로 변환하는 메서드

- parseInt() : 정수 형태의 변환
- parseFloat() : 실수 형태로 변환

## 연산자

### 산술 연산자

- 사칙연산
  - `+-*/%`
    - 문자의 사칙연산
      - 더하기 연산은 문자열 추가
      - 나머지 연산은 숫자로 계산

### 증감 연산자

```js
    var num = 10;

    console.log(++num); // num+1 후 num 출력
    console.log(--num); // num-1 후 num 출력
    console.log(num++); // num 출력 후 num + 1
    console.log(num--); // num 출력 후 num - 1
```

### 비교 연산자

- `==` : 값이 같다
- `===` : 데이터 타입과 값이 같다
- `!==` : 값이 같지 않다

- `>`
- `>=`
- `<`
- `<=`

### 논리 연산자

- `&&` : and
- `||` : or

## 조건문

### if 문

```js
    var a = 20;
    var b = 40;
    var c = 60;

    if (a > b) { console.log("a는 b보다 크다");
    } else if (b > c) { console.log("b는 c보다 크다");
    } else if (a < c) { console.log("a는 c보다 작다");
    } else if (b < c) { console.log("b는 c보다 작다");
    } else { console.log("모든 조건을 만족하지 않는다");
    } 
```

```js

    }
```
