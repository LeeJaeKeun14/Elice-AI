## Numpy

```python
A = np.array([[1, 2], [3, 4]])

print(A ** 2)
# [[ 1  4  ]
#  [ 9  16 ]]
print(3 ** A)
# [[ 3  9  ]
#  [ 27 81 ]]
print(A * A)
# [[ 1  4  ]
#  [ 9  16 ]]
print(np.dot(A, A)) # A*A 와 다르다
# [[ 7  10 ]
#  [ 15 22 ]]
print(A == A)
# [[ True True ]
#  [ True True ]]
a = np.array([1, 1, 0, 0], dtype = bool)
b = np.array([1, 0, 1, 0], dtype = bool)

print(np.logical_or(a, b))
# [True, True, True, False]
print(np.logical_and(a, b))
# [True, False, False, False]

print(A.sum()) # 10
print(A.min()) # 1
print(A.max()) # 4
print(A.argmin()) # 0   # 최소/최대값의 인덱스를 반환
print(A.argmax()) # 3
 
print(np.all(a)) # False  # array 내 모든 값이 True 인가
print(np.all(b)) # False
print(np.any(a)) # True   # array 내 하나라도 값이 True 인가
print(np.any(b)) # True

print(A.mean())     # 2.5    np 든 A. 이든 상관 없음
print(np.median(A)) # 2.5    np에만 median이 있음
print(A.std())      # 1.118  np 든 A. 이든 상관 없음
```
