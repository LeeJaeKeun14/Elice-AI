## 2. 문자열 함수

### 문자열 함수 : startswith()

```python
word = "superman"
print(word.startswith("s")) # True

if word.startswith('a'):
    print("a로 시작하는 단어입니다")
```

**속도는 인덱싱이 더 빠르다**

- 하지만 우리가 startswith를 사용해야 하는 이유는 가독성(1년뒤 코드를 읽더라도 한눈에 첫글자 비교를 위한 코드구나를 알 수 있음)

### 문자열 함수 : split()

```python
S = "my name is elice"
print(S.split())
# ["my", "name", "is", "elice"]

S = "사과,귤,배,바나나"
print(S.split(","))
```

```python
numbers = "    1   2 3    "
print(numbers.split()) # ["1", "2", "3"]

print(numbers.split(" ")) # ["", "", "1", "", "2", "", "3", ...]
```

- 대표적인 공백문자
  - ` ` : 빈칸(스페이스바)
  - `\t` : Tab(Tab키)
  - `\n` : Newline(엔터 키)

### 문자열 인 척 하는 리스트 함수 : append()

```python
numbers = []
numbers.append(1)
print(numbers) # [1]

numbers = [1, 2, 3, 4]
ls = []

for n in numbers:
    if n < 2:
        ls.append(n)
print(ls) # [1]
```

### 문자열 함수 : lower()

```python
intro = "My name is Elice!"
print(intro.upper()) # 'MY NAME IS ELICE!'
print(intro.lower()) # 'my name is elice!'
```

### 문자열 함수 : replace()

```python
intro = '제 이름은 Elice입니다."
print(intro.replace("Elice", "엘리스"))
# "제 이름은 엘리스입니다."
```

# 영어 단어 모음으로 시작하는 텍스트 파일 분석

## 파일 다루기

### 파일 열기 닫기

```python
file = open('data.txt')
content = file.read()
file.close()
```

- 파일 자동으로 닫기

```python
with open('data.txt') as file:
    content = file.read()
```

- 줄 단위로 읽기

```python
contents = []
with open('data.txt') as file:
    for line in file:
        content.append(line)
```

- 파일의 모드

```python
with open('data.txt', 'w') as file:
    file.write('Hello')
```

## 3. 리스트로 리스트 만들기

```python

    result.append(i[0])
```

- list comprehension

```python
ls = ["10", '20', '30']
result = [i[0] for i in ls]
```

```python
ls = [1, 3, 5, 7]
result = [n + 1 for n in ls]
```

- 특정 원소 걸러내기

```python
ls = [1, 3, 4, 5, 6, 7]
result = [n + 1 for n in ls if n % 2 == 0]
```
