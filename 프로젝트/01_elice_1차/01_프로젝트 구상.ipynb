## gitlab

- 일일회고(Wiki)
  - 일반적으로 해당 프로젝트에 대한 백과사전을 남기기 위한 목적으로 사용

- 오프스아워(Issue)
  - 일반적인 용법이 따로 없을 정도로 자유도가 높은 기능입니다.
  - 주로 request todo features, bugreporting, discussion등을 위해 사용합니다.
  - 주로 질문을 위해 사용
  - 프로젝트 진행 시, 어떠한 상황, 문제 또는 계획을 개발 할 예정이라는 것을 알려주는 이정표입니다.

- Merge Request(코드리뷰)
  - 일반적으로 PR(Pull request)로 알려져있습니다.
  - 대표적인 use case는 
    - dev branch -> master branch로의 병합 요청입니다.
  - MR하지 않는다면) local rapo에서
    - dev -> master merge 이후 remote로 push하는 방법도 있습니다.
  - 위와 같은 방법을 사용하지 않고 MR이라는 복잡한 방법을 사용하는 이유는...

- 프로젝트 구조
  - 1. master Branch
  - 2. Dev Branch
  - 3. Feature Branch

- README.md
  - 포함되어야 하는 내용
    - 구성 안내, 기술 스택 등
    - 프로젝트 기능 설명
    - 프로젝트 구성도
    - 프로젝트 팀원 역활 분ㄴ담
    - 버전


- GitLab issue
  - 개발 중 어려웠거나 궁금한 내용을 매일 오후 6시까지 등록
  - Issue의 Assignee을 오피스아워 진핸자 코치님 태크
