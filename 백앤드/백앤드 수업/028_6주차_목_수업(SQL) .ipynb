## 효율성의 원칙: 약속, 단순, 확장
- 약속(Promise)
  무엇을 주고, 무엇을 받을지 확실히 정의하라

- 단순(Simplicity)
  다양한 필요를 단순하게 제공하라
  문제를 재정의하면 쉬워진다
  마음속 카테고리를 바꾸면 문제가 쉬워진다
- 확장(Flexibility)
  고객의 다양한 필요는 확장된다.
  문제의 규모도 확장된다. 
  변화에 대비하라

### KEY

- Key 는 컬럼에 적용되는 속성으로 데이터의 무결성을 위해 사용됩니다


1. 기본 키(Primary Key, PK)
  - 기본 키는 한 테이블 내에서의 여러 레코드 중에서도 하나를 특정 할 수 있게하는 조건입니다.
    - 예를 들면, 중복되지 않는 닉네임, 숫자, 등등
  - 특정할 수 있다는 것이 가장 중요한 성질이기 때문에 중복되지 않고, NULL일 수가 없다는 특징을 가집니다.
  - 또한 한 테이블에는 하나의 기본 키만 가질 수 있습니다.
    - 기본키(id, title)

![](image/028_001.png)

2. 유니크 키(Unique Key)
  - 중복되는 값을 가지지 않겠다는 조건입니다.
  - 기본 키와 비슷해보이지만 NULL이 될 수 있고 테이블에서 여러 개를 가질 수 있다는 것이 차이점입니다.

3. 외래 키(Foreign Key)
  - 한 테이블에서 다른 테이블의 레코드를 특정할 수 있도록 하는 조건입니다.
  - 다른 테이블의 레코드를 특정할 수 있어야 하므로 다른 테이블의 기본 키인 경우가 많습니다.

  - 지난 강의 때 JOIN
  - user_id와 buyer_id로 각각 customer와 orderInfo 테이블을 연결하는 문제였습니다.
  - buyer_id로 customer에서 하나의 레코드를 특정할 수 있으므로 buyer_id는 외래 키가 될 수 있습니다.

  - 하지만 customer 테이블에서 엘리스가 자신의 user_id를 ‘토끼’로 바꿨다고 생각해봅시다.
  - userInfo 테이블에서는 여전히 buyer_id가 ‘엘리스’이기 때문에 데이터의 정합성이 깨집니다.
  - 따라서, 외래 키는 한 테이블에서 레코드가 변경될 경우 자동으로 다른 테이블에서도 변경되도록  하는 기능을 제공합니다.

## 정규화

- 정규화의 종류에는 1NF, 2NF, 3NF, BCNF, 4NF, 5NF, 6NF 로 모두 7가지가 있지만 우리가 필수로 알아야 하는 정규화는 3NF까지 입니다.
- **제1 정규형(1NF)**은 테이블에서 각 레코드들이 하나의 값만 같도록 하는것입니다.
- **제2 정규형(2NF)**은 간단히 설명하자면 특정 컬럼이 기본 키의 부분집합에 종속되면 안된다는 것입니다.
  - 단, 여기서 기본 키는 두 개 이상으로 이루어진 복합 키입니다.
- **제3 정규형(3NF)**은 테이블 내에 존재하는 이행적 종속을 제거하는 것입니다.
  - 이행적이라는 것은 A→B이고 B→C일 때, A→C인 걸 의미합니다. 이 때, C는 A에 이행적 종속입니다.

- 사실 많은 개발자들이 정규화라는 개념을 외우기보다는 체득한 경험을 토대로 설계하는 경우가 많습니다.
