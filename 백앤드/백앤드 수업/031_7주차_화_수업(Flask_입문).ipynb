## **MVC**

- M: Model
  - 모델은 어플리케이션에서 사용될 데이터를 정의한 것을 말합니다.
  - 일반적으로는 데이터베이스 테이블과 대응되도록 만듭니다.

- V: View
  - 뷰는 사용자에게 화면으로 보여지는 모든 것을 말합니다.
  - 사용자는 뷰를 통해 데이터를 제어할 수 있으며, 뷰에서는 데이터를 저장하지 않습니다.

- C: Controller
  - 컨트롤러는 모델과 뷰를 연결해주는 역활을 합니다
  - 클라이언트의 요청을 받아 모델로 넘겨주거나, 모델에서 받은 데이터를 뷰로 넘겨주는 역활 등이 있습니다.

- 간단하면서도 효율적인 장점들을 제공하기 때문에 많은 웹 프레임워크들이 MVC 패턴을 채용하여 만들어졌습니다.
  - RAILS
  - Laravel
  - spring

- 플라스크는 마이크로 웹 프레임워크이기에 기본적인 기능만 제공됩니다.
  - 따라서 플라스크에서 MVC 패턴을 사용할지는 개발자의 자유!
