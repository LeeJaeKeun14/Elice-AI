### Viewport

```html
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
```

- 뷰포트의 넓이는 디바이스 넓이, 비유을 1배

### position

1. 부모 자식 지간에 발생하는 마진 병합 현상 유무(2차원 O, 3차원 X)
2. top, left, right, bottom(2차원 X, 3차원 O)
3. 자식의 높이값이 부모의 높이값에 영향을 주는지 유무(2차원 O, 3차원 X)

- position: static
  - 위 속성 OXO
- position: fixed
  - 위 속성 XOX
- position: relative
  - 위 속성 OOO
- position: absolute
  - 위 속성 XOX

### z-index

- 높이를 구분하는 값

### background-size

- background-size: contain
  - 사진에 맞추고, 여백발생

- cover
  - 화면에 맞추고, 사진이 짤림

### background-attachment

- background-attachment: scroll;
  - 배경, 내용이 동시에 이동

- background-attachment: fixed;
  - 배경 이미지는 고정, 내용이 이동

### y축 중앙 정렬 방법

```css
    {top: 50%; 
    transform: translateY(-50%);}
```
top : 시작점이 50% 하지만 박스의 크기만큼 차이가 있음
transform: translateY(); : 크기의 절반만큼 올라가면 정 중앙이 전체의 중앙에 위치

### Full Clip 플러그인

- 용량이 작다
  - https://gitgub.com/bdimitrovski/full_clip
  
```html
  <script src="js/jquery-3.3.1.js"></script>
  <script src="js/fullclip.js"></script>
  <script>
    $(".fullBackground").fullClip({
        images: ['img/1.png', 'img/2.png', 'img/3.png'],
        transitionTime: 1000,
        wait: 3000
    })
  </script>
```

### input

- 단문 기입
- label 테그와 input태그가 같이 사용됨

```html
    <label for="topic">제목</label>
    <input type="text" id="topic" name="topic>
```

- 체크박스
  - 중복 가능

```html
    <label for="apple">
        <input type="checkbox" id="apple" name="apple">사과
    </label>
    <label for="banana">
        <input type="checkbox" id="banana" name="banana">바나나
    </label>
```

- 라디오
  - 중복 불가
  - 
```html
    <label for="male">
        <input type="radio" id="male" name="gender">남자
    </label>
    <label for="female">
        <input type="radio" id="female" name="gender">남자
    </label>
```
