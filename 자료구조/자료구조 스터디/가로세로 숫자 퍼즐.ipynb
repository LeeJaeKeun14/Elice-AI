## 가로세로 숫자 퍼즐

각 칸에 0 이상 100 미만의 두 자리 숫자로 이루어진 정수가 적혀있는 N * N 크기의 격자가 있습니다.
(10 미만의 수는 빈자리를 0으로 표기합니다.)

퍼즐의 답은 가로 방향, 세로 방향, 대각선 방향으로 연속된 4개의 수의 곱 중 가장 큰 값입니다.

예를 들어 5 * 5 크기의 퍼즐이 있을 때,

> 16 39 05 42 96
> 
> 86 56 00 48 35
> 
> 19 80 81 68 05
> 
> 04 52 08 83 97
> 
> 88 36 68 87 16

퍼즐의 답은 48 x 68 x 83 x 87 = 23,569,344이 됩니다.

N * N 크기의 퍼즐이 주어졌을 때, 퍼즐의 답을 구하는 프로그램을 작성하세요.

***

### [입력]
첫째 줄에 숫자 퍼즐의 크기를 나타내는 자연수 N을 입력합니다.
(4<=N<=20)(4<=N<=20)
(4<=N<=20)
둘째 줄부터 N개의 줄에는 퍼즐을 구성하는 수를 공백으로 구분하여 입력합니다. 수의 범위는 0 이상 100 미만이며 10 미만의 수는 빈자리를 0으로 표기하여 두 자리 숫자로 입력됩니다.

### [출력]
가로, 세로, 대각선 방향으로 연속된 4개의 수의 곱 중 최댓값을 출력합니다.
[입력 예시]
```
    5
    16 39 05 42 96
    86 56 00 48 35
    19 80 81 68 05
    04 52 08 83 97
    88 36 68 87 16
```

[출력 예시]
```
    23569344
```

````python
from queue import Queue
    
N = int(input())

Map = []
for _ in range(N):
    Map.append(list(map(int, input().split())))

max_num = 0  # 최대값


def calculate(max_num, q, result, count, zero, current):
    q.put(current)

    if current == 0: zero = True
    else: result *= current

    if count > 3: 
        get_num = q.get()
        if not zero:
            max_num = max(max_num, result)

        if get_num == 0:
            zero = False
        else:
            result /= get_num
    return max_num, q, result, count, zero
    

# 가로
for i in range(N):
    q = Queue()
    result, count = 1, 0
    zero = False

    for j in range(N):
        count += 1
        
        max_num, q, result, count, zero = calculate(max_num, q, result, count, zero, Map[i][j])

# 세로
for j in range(N):
    q = Queue()
    result, count = 1, 0
    zero = False
    
    for i in range(N):
        count += 1
        
        max_num, q, result, count, zero = calculate(max_num, q, result, count, zero, Map[i][j])
        
## 대각선

# for i in range(-N, N):
#     print()
#     for j in range(N):
#         y, x = j, i + j
#         if -1 < x < N:
#             print(j, i + j)

# for i in range(0, N * 2):
#     print()
#     for j in range(N):
#         y, x = j, i - j
#         if -1 < x < N:
#             print(y, x)


# 좌상 우하
for i in range(-N, N):
    q = Queue()
    result, count = 1, 0
    zero = False
    for j in range(N):
    
        count += 1
        y, x = j, i + j
        if -1 < x < N:
            max_num, q, result, count, zero = calculate(max_num, q, result, count, zero, Map[y][x]) # y, x
            

# 우상 좌하
for i in range(0, N * 2):
    q = Queue()
    result, count = 1, 0
    zero = False
    for j in range(N):
        count += 1
        y, x = j, i - j
        if -1 < x < N:
            max_num, q, result, count, zero = calculate(max_num, q, result, count, zero, Map[y][x]) # y, x

print(int(max_num))
````
